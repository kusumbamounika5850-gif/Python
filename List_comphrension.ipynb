#List Comphrension

```python
cubes = []

for i in range(1, 11):
    cubes.append(i**3)

print(cubes)
```

```python
#new_list = [expression for_loop]
cubes = [i**3 for i in range(1, 11)]
print(cubes)
```

```python
mylist = []

for i in range(1, 6):
    mylist.append(i)

print(mylist)
```

```python
mylist = [i for i in range(1, 6)]
print(mylist)
```

```python
mylist = [3 for i in range(1, 6)]
print(mylist)
```

```python
even = []

for num in range(1, 11):
    if num % 2 == 0:
        even.append(num)

print(even)
```

```python
#new_list = [expression for_loop if condition]
even = [num for num in range(1, 11) if num % 2 == 0]
print(even)
#new_list = [expression1 if condition else expression2 for_loop]
```

```python
result = []

for num in range(1, 11):
    if num % 2 == 0:
        result.append(num ** 2)
    else:
        result.append(num ** 3)
print(result)
```

```python
result = [num ** 2 if num % 2 == 0 else num ** 3 for num in range(1, 11)]
print(result)
```

```python
list1 = [1, 2, 3, 4]
list2 = [1, 20, 30, 4]
result = []

for x in list1:
    for y in list2:
        if x == y:
            result.append(y)

print(result)
```
