```python
# Find the given number is perfect or not:

n = 6
t = []

for i in range(1,n):
  if n % i == 0:
    t.append(i)

print(t)

if sum(t) == n:
  print(f'{n} : is a Perfect Number')

else:
  print(f'{n} : is not a perfect number')
```

```python
# find the given number is palindrome or not:

f = 121
t = f
s = 0


while f > 0:
  res = f % 10
  s = (s * 10) + res
  f = f // 10


if t == s:
  print(f'Palindrome')
else:
  print(f'not a palindrome')
```

```python
# find the given number is Amstrong or not:

f = 153
t = f
s = 0 # 153

while f > 0:
  res = f % 10 # 1
  s = s + (res * res * res)
  f = f // 10 # 0

if t == s:
  print(f'Amstrong')
else:
  print(f'not a Amstrong')
```

```python
# Find the given number is strong number or not:

import math

f = 145
t = f
s = 0 # 145

while f > 0:
  res = f % 10 # 1
  s = s + math.factorial(res)
  f = f // 10 # 0

if t == s:
  print(f'Strong')
else:
  print(f'not a Strong')
```
