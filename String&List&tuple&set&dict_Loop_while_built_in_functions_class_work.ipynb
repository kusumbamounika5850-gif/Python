#**Slicing:

Python follows ( n-1 ) syntax:









```python
#Slicing

p = 'Kusumba Mounika'

print(p[2:7])
print(p[8:14])

# using backward
print(p[-1:-8])
```

```python
p = 'Kusumba Mounika'


# using backward
print(p[-11:-2])
print(p[-15:-1])
print(p[:9])
print(p[:])
```
