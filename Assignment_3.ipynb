```python
n = 8
c = 0

for i in range(1,n + 1):
  if n % i == 0:
    c = c + 1


if c == 2:
  print(f'{n} : is a prime number')
else:
  print(f'{n} : not a prime number')
```

```python
for i in range(2,101):
  c = 0
  for j in range(1,i + 1):
    if i % j == 0:
      c = c + 1

  if c == 2:
    print(f'{i} : is a prime number')
  else:
    print(f'{i} : not a prime number')
```

```python
t = 'data sCience 12444'
upper = 0
lower = 0
sp = 0
num = 0

for i in t:
  if i.isupper():
    upper = upper + 1
  elif i.islower():
    lower = lower + 1
  elif i.isnumeric():
    num = num + 1
  else:
    sp = sp + 1

print(f'Upper letter: {upper}')
print(f'Lower letter: {lower}')
print(f'Space : {sp}')
print(f'Numerical values : {num}')
```
