```python
#5.Write a Python program to count Uppercase, Lowercase, special character and numeric values in a given string.
#sample input : data sCience 12444
#sample output :
#upper case : 1
#lower case : 10
#numbers : 5
#space : 2
a = 'data sCience 12444'
low=0
upper=0
numbers=0
space=0
for i in a:
  if i.isupper():
    upper=upper+1
  elif i.islower():
    low=low+1
  elif i.isnumeric():
    numbers=numbers+1
  else:
    space=space+1
print(f'upper case:{upper}')
print(f'lower case:{low}')
print(f'numbers:{numbers}')
print(f'space:{space}')
```
