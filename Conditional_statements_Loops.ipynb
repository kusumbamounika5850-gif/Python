```python
#frequency checking.
x = 'python is easy programming language'
y = x.split()
print(y)
b = {}
for i in x:
  if i not in b:
   b[i]=1
  else:
    b[i]=b[i]+1
print(b)
c = max(b.values())
print(c)
for j in b:
  if b[j]==c:
    print('most frequent character:',j)
else:
    print('frequency:',c)


```
