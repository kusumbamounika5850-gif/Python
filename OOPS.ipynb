```python
class PRIME:
  def calculation(self,a,b):
    for i in range(a,b): # (1,101)
      c = 0
      for j in range(1,i+1):
        if i % j == 0:
          c = c + 1
      if c == 2:
        print(f'Prime : {i}')
      else:
        print(f'Not Prime : {i}')

anil = PRIME()
anil.calculation(1,101)
```

**Inheritance**

- `Single Inheritance `
- `Multi Level Inheritance `
- `Multiple Inheritance `
- `Hierarichal Inheritance `
- `Hybrid Inheritance `

```python
# single Inheritance

class PARENT:
  def building(self):
    print(f'I have My own Buildng')


class CHILD(PARENT):
  def job(self):
    print(f'I have my own Job')


kumar = CHILD()
kumar.job()
kumar.building()
```

```python
# Multi Level Inheritance

class GRANDFATHER:
  def shopping_mall(self):
    print(f'I have My own Shopping Mall')

class FATHER(GRANDFATHER):
  def building(self):
    print(f'I have My own Buildng')


class CHILD(FATHER):
  def job(self):
    print(f'I have my own Job')

anil = CHILD()
anil.job()
anil.building()
anil.shopping_mall()
```

```python
# Multiple Inheritance

class MOTHER:
  def shopping_mall(self):
    print(f'I have My own Shopping Mall')

class FATHER:
  def building(self):
    print(f'I have My own Buildng')


class CHILD(FATHER,MOTHER):
  def job(self):
    print(f'I have my own Job')

anil = CHILD()
anil.job()
anil.building()
anil.shopping_mall()
```

```python
# Hierarichal Inhertiance

class FATHER:
  def shopping_mall(self):
    print(f'I have My own Shopping Mall')

class B1(FATHER):
  def building(self):
    print(f'I have My own Buildng')


class B2(FATHER):
  def job(self):
    print(f'I have my own Job')

class B3(FATHER):
  def PROPERTY_(self):
    print(f'I have my own RICH PROPERTY')


anil = B3()
anil.PROPERTY_()
anil.shopping_mall()
```

```python
def fun(a = None,b = None):
  if a == None and b == None:
    print('I am from USA')
  elif a == None and b != None:
    print('I am from UK')
  elif a != None and b == None:
    print('I am from India')
  else:
    print('I am somewhere')


fun() # calling
fun(a = 10)
fun(b = 10)
fun(10,20)
```

```python
# Abstraction
from abc import ABC,abstractmethod

class WHATSAPP_VERSION_1(ABC):
  def chat(self):
    print('chat available')

  @abstractmethod
  def status(self):
    pass


class WHATSPP_VERSION_2(WHATSAPP_VERSION_1):
  def status(self):
    print(f'New changes ')


obj = WHATSPP_VERSION_2()
obj.status()
```
