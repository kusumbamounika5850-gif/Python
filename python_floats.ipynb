#Python floats

```python
x = 1.1
y = 1.5
z = -3.59

print(type(x))
print(type(y))
print(type(z))
```

```python
x = 54e5
y = 21E5
z = -87.7e100

print(type(x))
print(type(y))
print(type(z))
```

```python
x = 6+8j
y = 9j
z = -3j

print(type(x))
print(type(y))
print(type(z))
```

```python
x = 2    # int
y = 2.4  # float
z = 3j   # complex

#convert from int to float:
a = float(x)

#convert from float to int:
b = int(y)

#convert from int to complex:
c = complex(x)

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

```python
x = int(3)   # x will be 1
y = int(5.8) # y will be 2
z = int("7") # z will be 3
print(x)
print(y)
print(z)
```

```python
x = str("s1") # x will be 's1'
y = str(2)    # y will be '2'
z = str(3.0)  # z will be '3.0'
print(x)
print(y)
print(z)
```
