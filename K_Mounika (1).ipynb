```python
a = [5, 6, 5, 7, 7, 8]
target_sum = 13
b = []
for item in a:
 if item not in b:
  b.append(item)
found_pairs_list = []
for i in range(len(b)):
    for j in range(i + 1, len(b)):
      if b[i] + b[j] == target_sum:
        found_pairs_list.append(f"[{b[i]},{b[j]}]")
        found_pairs_list.append(f"[{b[j]},{b[i]}]")
        pairs_count = len(found_pairs_list)
#The final count of pairs is 4
print(f"Total pairs count: {pairs_count}")
print(f"Found pairs: {', '.join(found_pairs_list)}")
```
